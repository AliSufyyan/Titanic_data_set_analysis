# Classwise Survivel
## Key Findings:
### First class has the highes survivel rate

```python
survival_count=df.groupby(['class', 'survived'], observed=False).size().unstack()
```

```python
survival_count.plot(kind='bar', stacked=True, color=['green','red'])
plt.title('Survivel  Rate By Class')
plt.legend(['Survived', 'Not Survived'])
```

# Survivel Rate
## Key Findings:
### Number of not survived passengers is higher in each class 

```python
sex_survivel=df.groupby(['sex', 'survived']).size().unstack()
```

```python
sex_survivel.plot(kind='bar')
plt.title('Gender wise Survivel')
plt.legend(['Survived', 'Not Survived'])
```

# Ports
## key Findings:
### Most of  the passengers are boarded from the port S: Southampton, England

```python
bin_count=[0,12,18,60,100]
label_name=['Childeren', 'Young Adults','Adults','Seniors']
df['Age_Group']=pd.cut(df['age'], bins=bin_count, labels=label_name, right=False)
```

```python
age_group_survivel_rate=df.groupby('Age_Group',observed=False)['survived'].mean()*100
```

```python
sns.barplot(x=age_group_survivel_rate.index, y=age_group_survivel_rate.values,hue=age_group_survivel_rate.index, palette='Set3')
```
